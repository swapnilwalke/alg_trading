Import libraries ---

```python
import numpy as np
import pandas as pd
import yfinance as yf
import math
from scipy import stats
import xlsxwriter
from yahooquery import Ticker
```

```python
stocks = pd.read_csv('./resources/nifty50.csv')
col_symbols = stocks['Symbol'].tolist()
yahoo_ticker_appender=".NS"
yahoo_ticker_col = [f"{i}{yahoo_ticker_appender}" for i in col_symbols]
stocks['YTicker']=yahoo_ticker_col
stocks =stocks.drop(index=0)
# stocks=stocks.drop(['Change','%Change','Traded Volume(lacs)','Traded Value(crs)','52 Week High','52 Week Low','365 Days % Change','30 Days % Change'],axis=1)
# print(stocks['YTicker'])
```

```python
my_columns = ['Ticker','Price','One Year Price Return','Number of Shares to Buy']
final_dataframe = pd.DataFrame(columns=my_columns)
# print(yf.Ticker('ACC.NS').fast_info)
for stock in stocks['YTicker']:
    stock_ticker = stock
    stock_info = yf.Ticker(stock)
    stock_one_year_price_return = stock_info.fast_info['yearChange']
    stock_last_price=stock_info.fast_info['last_price']
    indexOfLastRow = len(final_dataframe.index)
    final_dataframe.loc[indexOfLastRow]=[stock_ticker,stock_last_price,stock_one_year_price_return,'N/A']
final_dataframe
```

Remove Low Momentum Stocks...


```python
final_dataframe.sort_values('One Year Price Return', ascending=False, inplace=True)
final_dataframe.reset_index(inplace=True)
final_dataframe
```
